This project uses two datasets provided by the UBC research group studying a Minecraft server.  
The first dataset, players.csv, contains 196 rows and 7 columns, with one record per player.  
It includes variables such as experience (categorical), subscribe (TRUE/FALSE),  
played_hours (numeric), gender, and Age.  
There are 144 subscribers and 52 non-subscribers, and Age includes two missing values.  
played_hours is strongly right-skewed, ranging from 0 to 223.1 hours.  
Overall, this dataset describes who the players are and their total engagement level.

(2) Questions

**Broad question:**  
What player characteristics and behaviors are most predictive of subscribing to the game-related newsletter?

**Specific predictive question:**  
Can a player’s experience level, total played hours, gender, and age predict whether they subscribe to the newsletter?


(3) Exploratory Data Analysis and Visualization:

```r
players |>
  group_by(experience) |>
  summarise(mean_played = mean(played_hours, na.rm = TRUE))
```

```r
ggplot(players, aes(x = experience, y = played_hours, fill = experience)) +
  geom_boxplot() +
  labs(title = "Played Hours by Experience Level",
       x = "Experience Level", y = "Total Played Hours")
```

```r
players |>
  group_by(experience) |>
  summarise(prop_subscribe = mean(subscribe)) |>
  ggplot(aes(x = experience, y = prop_subscribe, fill = experience)) +
  geom_col() +
  labs(title = "Newsletter Subscription Rate by Experience Level",
       x = "Experience Level",
       y = "Proportion Subscribed")
```

```r
ggplot(players, aes(x = Age)) +
  geom_histogram(binwidth = 2, fill = "red", color = "blue") +
  labs(title = "Distribution of Player Ages",
       x = "Age", y = "Count")
```

The exploratory plots show clear differences in engagement across experience groups.  
Most players have very low total playtime, but a small number of “Pro” and “Veteran” players accumulate extremely high played hours, creating a strong right-skew.  
Subscription behaviour appears related to this pattern: players with higher experience levels tend to have longer play histories and are more likely to subscribe.

The age distribution is concentrated between 15 and 25, with only a few older players.  
This suggests that the player base is young overall, and age may have limited variation as a predictor.

These patterns indicate that experience level and total played hours are likely to be informative predictors in modelling subscription behaviour.

(4) Methods and Plan：
To address the research question, I plan to use a classification model to predict whether a player subscribes to the newsletter based on their experience level, total played hours, age, and gender.  
A logistic regression model will first be applied for interpretability, followed by k-nearest neighbours (k-NN) to capture potential nonlinear relationships.  
Data will be pre-processed by encoding categorical variables, scaling numeric predictors, and merging player and session data using `hashedEmail`.  
The data will be split into 70% training and 30% testing sets, with 5-fold cross-validation for parameter tuning.  
Potential limitations include class imbalance and the small sample size (n = 196), which may affect generalizability.
